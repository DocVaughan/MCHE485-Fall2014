## Problem 1

```python
m = 1.0                     # kg
k = (1*2*np.pi)**2          # N/m
c = 0.5                     # Ns/m

wn = np.sqrt(k / m)         # natural frequency (rad/)
z = c / (2*m*wn)            # damping ratio
wd = wn * np.sqrt(1 - z**2) # damped natural frequency
```

### Problem 1.d.

```python
x0 = [1.0, 5.0]             # initial conditions [x0, v0]

t = np.linspace(0,5,501)
x = np.exp(-z*wn*t)*(x0[0]*np.cos(wd*t) + (z*wn*x0[0] + x0[1])/wd * np.sin(wd*t))
```

### Problem 2.d.ii.

```python
# Define the matrices - We'll use the values from the book example
m1 = 10.0
m2 = 1.0
mf = 100.0

k = 100.0 
kf = 1000.0

c = 3.0

e = 0.1

M = np.asarray([[m1+m2, 0],
                [0,    mf]])

K = np.asarray([[k,      -k],
                [-k, k + kf]])

C = np.array([[c, -c],
              [-c, c]])
```

```python
w = np.linspace(0,6,1200)
X = np.zeros((len(w),2)) + 1j*np.zeros((len(w),2))

# This is (K-w^2 M)^-1 * (w^2*F)
for ii, freq in enumerate(w):
    F = [m2 * e * freq**2, 0]
    X[ii,:] = np.dot(linalg.inv(K  + 1j*freq*C - freq**2 * M), F)
```

```python
# Define the system as a series of 1st order ODES
def eq_of_motion(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x, x_dot]
        t :  time
        p :  vector of the parameters:
                  p = [m, k, c, mu, g]
    """
    x, x_dot = w
    m, k, c, mu, g = p

    # Create sysODE = (x', x_dot'):
    sysODE = [x_dot,
             (-k * x - mu*m*g*np.sign(x_dot) - c*x_dot) / m]
    return sysODE
```

```python
# Define the sytsem parameters
m1 = 4.0
m2 = 1.0
k = 1.0
c = 0.05

p = [m1, m2, k, c]
```

```python
# Set up simulation parameters - Same for parts d.i. and d.ii.

# ODE solver parameters
abserr = 1.0e-9
relerr = 1.0e-9
max_step = 0.01
stoptime = 20.0
numpoints = 10001

# Create the time samples for the output of the ODE solver
t = np.linspace(0.,stoptime,numpoints)
```

```python
# Define the system as a series of 1st order ODES
def eq_of_motion(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x1, x1_dot, x2, x2_dot]
        t :  time
        p :  vector of the parameters:
                  p = [m1, m2, k, c]
    """
    x1, x1_dot, x2, x2_dot = w
    m1, m2, k, c = p

    # Create sysODE = (x', x_dot'):
    sysODE = [x1_dot,
             -k/m1 * (x1-x2) - c/m1 * (x1_dot - x2_dot),
             x2_dot,
             k/m2 * (x1-x2) + c/m2 * (x1_dot - x2_dot) + f(t,p)/m2]
    return sysODE

def f(t, p):
    """
    defines the disturbance force input to the system
    """
    m1, m2, k, c = p
    
    f = 0.0 # no input
    
    return f
```
