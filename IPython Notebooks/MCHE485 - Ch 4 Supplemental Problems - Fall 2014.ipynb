## Problem 4.61

```python
# Define the known parameters
m1 = 1000.0         # kg
k1 = 36000.0        # N/m

F1 = 0.7 # N
F = [F1, 0]

w = 6.0   # rad/s

# Set up the range of m2 masses to look at
mass_range = np.linspace(0.1,1,200)

# Create an appropriately sized array of zeros to fill with amplitudes
x2 = np.zeros_like(mass_range)

# Now's let's see how |x2| changes with m2
for index, m2 in enumerate(mass_range):
    k2 = k1/m1 * m2    # N/m

    M = np.asarray([[m1, 0],
                    [0,  m2]])

    K = np.asarray([[k1 + k2, -k2],
                    [-k2,      k2]])
    
    # This is (K-w^2 M)^-1 * (w^2*F)
    X = np.dot(linalg.inv(K - w**2 * M), F)

    x2[index] = X[1]
```

```python
# We can also use a numerical solver to get a more exact solution

# Import the optimize module (We'll actually use a minimization function)
from scipy import optimize

def x2_amp_diff(m2):
    k1 = 36000.0
    m1 = 1000.0
    k2 = k1 / m1 * m2[0]    # N/m
    
    M = np.asarray([[m1, 0],
                    [0,  m2]])

    K = np.asarray([[k1 + k2, -k2],
                    [-k2,      k2]])
    
    # This is (K-w^2 M)^-1 * (w^2*F)
    X = np.dot(linalg.inv(K - w**2 * M), F)

    x2 = X[1]
    
    return x2 + 0.1

solution = optimize.fsolve(x2_amp_diff, 2)

print '\nThe second mass should be {:0.4f}kg.\n'.format(solution[0])
    
```

## Problem 4.63

```python
# Define the known parameters
m1 = 10.0         # kg
k1 = 4000.0        # N/m

F1 = 10.0          # N
F = [F1, 0]

w = 20.0           # input frequency (rad/s)

# Vibration Absorber case
m2 = 10.0          # kg
k2 = 4000.0        # N/m

M = np.asarray([[m1, 0],
                [0,  m2]])

K = np.asarray([[k1 + k2, -k2],
                [-k2,      k2]])

# This is (K-w^2 M)^-1 * (w^2*F)
X = np.dot(linalg.inv(K - w**2 * M), F)

print '\nIn this case, the magnitude of oscillation of the first mass is {}m.'.format(X[0])
print 'The magnitude of oscillation of the second mass is {}m.\n'.format(X[1])
```

```python
# Now, look at the "joined" case... the system reduces to a 1DOF mass-spring system
m = m1 + m2               # The new total mass (kg)

wn = np.sqrt(k1 / m)      # The natrual frquency of the 1DOF system

one_dof_amp = F1 / (m * (wn**2 - w**2))

print '\nThe amplitude of oscillation if the two masses are rigidly joined is {}m.\n'.format(one_dof_amp)
```

Looking at the two amplitudes, mass $m_2$ will have the same amplitude oscillation if used as a vibration absorber or rigidly attached to mass $m_1$.<br><br>

## Problem 4.68
In this problem, we can check that the only time that mass $m_3$ remains stationary, $m_2$ is also stationary.

```python
m1 = 3.0
m2 = 2.0
m3 = 1.0

k1 = 4.0
k2 = 4.0
k3 = 4.0

M = np.array([[m1, 0,  0],
              [0, m2,  0],
              [0,  0, m3]])

K = np.array([[ k1,      -k1,   0],
              [-k1, k1 + k2, -k3],
              [  0,     -k3,  k3]])

F2 = 1.0

F = np.array([0, F2, 0])

w = np.linspace(0.01,4,4000)
X = np.zeros((len(w),3))

# This is (K-w^2 M)^-1 * (w^2*F)
for ii, freq in enumerate(w):
    inverse = linalg.inv(K - freq**2 * M)
    X[ii,:] = np.dot(inverse, F)

# Let's mask the discontinuity, so it isn't plotted
pos = np.where(np.abs(X[:,0]) >= 10.0)
X[pos,:] = np.nan
w[pos] = np.nan
```
