We can also generate plots inline. To do this we issue the command:

```python
%matplotlib inline

# Import the plotting functions 
#  Note: Using the 'from module import *' notation is usually a bad idea. 
from matplotlib.pyplot import *
```

Now, let's try a simple plot.

```python
t = np.linspace(0, 5, 1000)  # Define the time vector (start,end,number of points)
y = np.sin(2 * np.pi * t)       
```

```python
plot(t, y)
xlabel(r'Time (s)', fontsize=22, labelpad=5)
ylabel(r'$\sin(2\pi t)$',fontsize=22, labelpad=8)
```

We can also issue commands to better format the plot and make the plot a little bit easier to read.

```python
fig = figure(figsize=(6,4))
ax = gca()
subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)
setp(ax.get_ymajorticklabels(),fontsize=18)
setp(ax.get_xmajorticklabels(),fontsize=18)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

xlabel('Time (s)',fontsize=22,weight='bold',labelpad=5)
ylabel('Position (m)',fontsize=22,weight='bold',labelpad=10)

plot(t, y,linewidth=2)
xlabel(r'Time (s)', fontsize=22, labelpad=5)
ylabel(r'$\sin(2\pi t)$',fontsize=22, labelpad=8)
```
