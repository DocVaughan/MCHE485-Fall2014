## Problem 1.14

```python
# Define the time array
t = np.linspace(0,2*np.pi,1000)         # Define an array of 1000 points between 0 and 2*pi

x = 2 * np.cos(t) + 0.05 * np.sin(10*t)
x_dot = -2 * np.sin(t) + 0.5 * np.cos(10*t)
x_ddot = -2 * np.cos(t) - 5.0 * np.sin(10*t)

# Let's also define the "no noise" version of each of these
x_noNoise = 2 * np.cos(t)
x_dot_noNoise = -2 * np.sin(t)
x_ddot_noNoise = -2 * np.cos(t)
```

## Problem 1.37

```python
# Let's use the symbolic toolbox Sympy (http://sympy.org) to solve this one.
import sympy

k, wn1, wn2 = sympy.symbols('k wn1 wn2')

wn1 = sympy.sqrt(k/100)
wn2 = sympy.sqrt(k/156.25)

k = sympy.solve(wn1 - wn2 - 2,k)

print '\nThe resulting spring constant, k, is {:0.2f} N/m.\n'.format(float(k[0]))
```

## Problem 1.54

```python
m = 0.06             # mass (kg)

tau = 0.5            # 0.5s period

wn = (2 * np.pi)/tau # calculate natural frequency from the period (rad/s)

k = m * wn**2        # solve for k

print '\nThe resulting spring constant is {:0.2f} N/m.\n'.format(k)
```
