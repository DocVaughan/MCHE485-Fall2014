## Problem 1
d. Plot the response, x(t), to initial conditions:

$\quad x(0) = x_0 \quad\mbox{and}\quad \dot{x}(0) = v_0 $

for damping ratios of $\zeta = 0.0$, $\zeta = 0.2$, and $\zeta = 0.7$. Be sure to clearly label the axes and differentiate between the responses.

e. Write the response, $x(t)$, to these same initial conditions.

<br>
The solution to the differential equation that we developed in class to plot the response. The solution for the underdamped case is:

$ \quad x(t) = e^{-\zeta\omega_nt}\left(a_1 e^{i \omega_d t} + a_2 e ^{-i \omega_d t}\right) $ 

*or*

$ \quad x(t) = e^{-\zeta\omega_nt}\left(b_1 \cos{\omega_d t} + b_2 \sin{\omega_d t}\right) $

To use this equation, we need to solve for $a_1$ and $a_2$ or $b_1$ and $b_2$ using the initial conditions. Here, let's use the sin/cosine form. Solving the equation for generic intial velocity, $\dot{x}(0) = v_0$, and a generic initial displacement, $x(0) = x_0$, we find:

$ \quad x(t) = e^{-\zeta\omega_nt}\left(x_0 \cos{\omega_d t} + \frac{\zeta \omega_n x_0 + v_0}{\omega_d} \sin{\omega_d t}\right) $

```python
# Define some parameters to generate the plot
t = np.linspace(0,2.5,500)
wn = 2*np.pi

# Define the initial conditions
x0 = [1.0, 10.0]


# zeta = 0.0
z = 0.0
wd = wn * np.sqrt(1 - z**2)

x_undamped = np.exp(-z*wn*t)*(x0[0]*np.cos(wd*t) + (z*wn*x0[0] + x0[1])/wd * np.sin(wd*t))


# zeta = 0.2
z = 0.2
wd = wn * np.sqrt(1 - z**2)

x_z0p2 = np.exp(-z*wn*t)*(x0[0]*np.cos(wd*t) + (z*wn*x0[0] + x0[1])/wd * np.sin(wd*t))


# zeta = 0.7
z = 0.7
wd = wn * np.sqrt(1 - z**2)

x_z0p7 = np.exp(-z*wn*t)*(x0[0]*np.cos(wd*t) + (z*wn*x0[0] + x0[1])/wd * np.sin(wd*t))
```

## Problem 3
This is a mass-spring-damper system. It first looks at step response, then at seismic, harmonic excitation.

The equation of motion for the system is:

$ \quad m \ddot{x} + c \dot{x} + kx = c \dot{y} + ky $

We could also write this equation in terms of the damping ratio, $\zeta$, and natural frequency, $\omega_n$.

$ \quad \ddot{x} + 2\zeta\omega_n\dot{x} + \omega_n^2x = 2\zeta\omega_n \dot{y} + \omega_n^2 y$

The general step response is described by:

$ \quad x(t) = c + e^{-\zeta \omega_n t}\left[a \cos(\omega_d t) + b \sin(\omega_d t)\right] $

The solution for zero initial condiations for the underdamped case is:

$ \quad x(t) = c + e^{-\zeta\omega_nt}\left(-c \cos{\omega_d t} - \frac{\zeta c}{\sqrt{1-\zeta^2}} \sin{\omega_d t}\right) $

```python
# Define the step input (We're using y_step, because we've already defined c as our damping coeff.)
y_step = 1.0

# Define some parameters to generate the plot
t = np.linspace(0,2.5,500)
wn = 2*np.pi


# zeta = 0.0
z = 0.0
wd = wn * np.sqrt(1 - z**2)

xstep_undamped = y_step + np.exp(-z*wn*t)*(-y_step*np.cos(wd*t) + (z*y_step)/(np.sqrt(1-z**2)) * np.sin(wd*t))


# zeta = 0.2
z = 0.2
wd = wn * np.sqrt(1 - z**2)

xstep_z0p2 = y_step + np.exp(-z*wn*t)*(-y_step*np.cos(wd*t) + (z*y_step)/(np.sqrt(1-z**2)) * np.sin(wd*t))


# zeta = 0.7
z = 0.7
wd = wn * np.sqrt(1 - z**2)

xstep_z0p7 = y_step + np.exp(-z*wn*t)*(-y_step*np.cos(wd*t) + (z*y_step)/(np.sqrt(1-z**2)) * np.sin(wd*t))
```

## Problem 4

The linearized equation of motion is:

$ \quad \ddot{\theta} + \omega_n^2 \theta = \frac{f}{m l} $

where:

$ \quad \omega_n = \sqrt{\frac{k - mgl}{m l^2}} $

If we look at the nonlinear version, with $f(t)=0$, we can see what moments must balance to create an equilibrium condition, $\theta_{eq}$.

$ \quad \ddot{\theta} + k \theta - m g l \sin{\theta} = 0 $

At equilibrium:

$ \quad k \theta = m g l \sin{\theta}$

Let's plot this relationship, to see what how the equilibrium angle behaves. Where the plots overlap are possible equilibrium angles.

```python
m = 1.0
l = 1.0
g = 9.81
k = 5.0

theta_eq = np.linspace(-np.pi, np.pi, 500)

spring_moment = k * theta_eq
grav_moment = m * g * l * np.sin(theta_eq)

# Set the plot size - 3x2 aspect ratio is best
fig = figure(figsize=(6,4))
ax = gca()
subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Change the axis units to serif
setp(ax.get_ymajorticklabels(),family='serif',fontsize=18)
setp(ax.get_xmajorticklabels(),family='serif',fontsize=18)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

# Define the X and Y axis labels
xlabel('Angle (deg)',family='serif',fontsize=22,weight='bold',labelpad=5)
ylabel('Moment',family='serif',fontsize=22,weight='bold',labelpad=10)

plot(theta_eq * 180/np.pi, grav_moment, linewidth=2, label=r'Gravity')
plot(theta_eq * 180/np.pi, spring_moment, linewidth=2, linestyle="--",label=r'Spring')

# uncomment below and set limits if needed
xlim(-180,180)
xticks([-180,-120,-60,0,60,120,180])
# ylim(0,10)
yticks([-20, -10, 0, 10, 20],['', '', '0', '', ''])

# Create the legend, then fix the fontsize
leg = legend(loc='upper right', ncol = 2, fancybox=True)
ltext  = leg.get_texts()
setp(ltext,family='serif',fontsize=18)

# Adjust the page layout filling the page using the new tight_layout command
tight_layout(pad=0.5)

# save the figure as a high-res pdf in the current folder
# It's saved at the original 6x4 size
# savefig('Midterm1_prob4d.pdf')

fig.set_size_inches(9,6) # Resize the figure for better display in the notebook
```
